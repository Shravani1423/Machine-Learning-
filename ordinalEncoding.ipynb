```python
import pandas as pd
```

```python
df =pd.DataFrame({"size" :["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head(3)
```

```python
# using sklearn method
```

```python
ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe =OrdinalEncoder(categories= ord_data)  #here we mention categories =ord_data becouse I wanted data in ord_data format
oe.fit(df[["size"]])
```

```python
df["size_en"]=oe.transform(df[["size"]])
df
```

```python
# using map function
```

```python
ord_data1 ={"s":0,"m":1,"l":2,"xl":3}
```

```python
df["size_en_map"]=df["size"].map(ord_data1)
df
```
